```python
housing_df = pd.read_csv(housing_data_file, sep='\t')
housing_df.describe()
```

```python
# Create dummy values
prop_type_dummies = pd.get_dummies(housing_df['PropertyType'])
```

```python
housing_df_2 = pd.concat([housing_df.drop('PropertyType', axis=1), prop_type_dummies], axis=1)
housing_df_2.head()
```

```python
housing_df_2['IsNewConstruction'] = np.where(housing_df_2['NewConstruction'] == True, 1, 0)
housing_df_2 = housing_df_2.drop('NewConstruction', axis=1)
```
