# Iterating

```python
# Row major interate read only
h = np.arange(16).reshape(4, 4)
for _h in np.nditer(h):
    print(_h)
```

```python
# Column major iteration read only
for _h in np.nditer(h, order='F'):
    print(_h)
```

```python
# More efficient
for _h in np.nditer(h, order='F', flags=['external_loop']):
    print(_h)
```

```python
# Use flag to make writable
h = np.arange(16)
print(h)
for _h in np.nditer(h, op_flags=['readwrite']):
    # Elipses notation means operate on the current element
    _h[...] = _h * _h
    

print(h)
```

```python
# Use flattening to iterate over all elements , by default ro major (all row elements together)
j = np.arange(16).reshape(4,4)
print(j)
print(j.flatten())
```

```python
# Flatten my column major
print(j.flatten(order='F'))
```

# Shallow copies using view

Chaning the original will change the view.
Reshaping a view doesn't change the original.

```python
fruits = np.array(['apple', 'banana', 'Mango', 'grape', 'watermelon'])
fruits_view1 = fruits.view()
print(fruit_view1)
```

```python
fruits_view1 is fruits
```

```python
fruit_view1.base is fruits
```

```python
# Deep Copy
```

```python
fruits_copy = fruits.copy()
```

```python
fruits_copy == fruits
```

```python
fruits_copy is fruits
```
